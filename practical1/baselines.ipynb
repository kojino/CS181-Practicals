```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd

X_train = pd.read_pickle('df_updated.pkl')
Y_train = X_train.gap
X_train = X_train.drop(['smiles','gap'],axis=1)
```

```python
X_test = pd.read_pickle('test_all.pkl')
```

```python
X_test = X_test.drop(['smiles','Id'], axis=1)
```

```python
from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor

ETR = ExtraTreesRegressor(n_estimators=50,max_features=6)
print "start fitting"
ETR.fit(X_train, Y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
print ETR.feature_importances_
```

```python
importances = ETR.feature_importances_
std = np.std([tree.feature_importances_ for tree in ETR.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X_test.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

```python
# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X_test.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X_test.shape[1]), indices)
plt.xlim([-1, X_test.shape[1]])
plt.show()
```
